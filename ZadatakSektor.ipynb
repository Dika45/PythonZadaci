```python
def listaBandOne(df):
    lis = {}
    for i in range(len(df)):
        if df.loc[i, 'band'] == 1:
            if df.loc[i, 'cell_id'][0] in lis:
                lis[df.loc[i, 'cell_id'][0]].append(df.loc[i, 'azimuth'])
            else:
                lis[df.loc[i, 'cell_id'][0]] = [df.loc[i, 'azimuth']]
                
        
        
    return lis
```

```python
def calctulationPoints(lis):
    l = []
    for idx, val in enumerate(lis):
        try:
            if lis[idx] > lis[idx+1]:
                prom = lis[idx]+(((lis[0]+360)-lis[idx])/2)
                l.append(prom)
            else:
                prom = lis[idx]+((lis[idx+1]-lis[idx])/2)
                l.append(prom)
        except:
            continue
            
    return l
        
```

```python
def calcSector(df, dic):
    for index, row in df.iterrows():
        if len(row.cell_id) == 2:
            df['sector'][index] = int(row.cell_id[1])
        else:
            nes = []
            for i in dic[row.cell]:
                if i >= row.azimuth:
                    si = i - row.azimuth
                    if si > 180:
                        si = 360 - si
                    nes.append(si)
                else:
                    si = row.azimuth - i
                    if si > 180:
                        si = 360 - si
                    nes.append(si)
                    
            ind = nes.index(min(nes))
            df['sector'][index] = ind+1
```

```python
calcSector(df, dic)
```

```python
def maxBand(df):
    dm = df.groupby(['cell', 'band']).agg({'cell_id': 'count'})
    
    for i in range(len(dm)):
        if dm['cell_id'].index[i][1] == 3:
            dm['cell_id'][i] = dm['cell_id'][i] / 2
    
    ki = {}
    li = 0
    for index, row in dm.iterrows():    
        if index[0] not in ki.keys():
            li = row.cell_id
            ki[index[0]] = [index[1],li]
        else:
            if li < row.cell_id:
                li = row.cell_id
                ki[index[0]] = [index[1],li]        
        
    
    return ki
```

```python
ban = maxBand(df)
```

```python
for key, value in ban.items():
    print('Vrsta celije: {} najveci band: {}  i ima {} celije u tom bendu'.format(key,value[0],value[1]))
```
